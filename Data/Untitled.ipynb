```python
from db_login import db_cursor
from db_info import DB_Info
import pandas as pd
import matplotlib.pyplot as plt

def load_data(table):
    mydb, cursor = db_cursor()
    db_info = DB_Info()
    cursor.execute("USE sports_betting;")

    sql = f"SELECT * FROM {table};"
    cursor.execute(sql)
    data = cursor.fetchall()
    cols = db_info.get_cols(table)
    return pd.DataFrame(data, columns=cols)
```

```python
df = load_data("ESPN_Games_NFL")
print(df.columns)
```

```python
for col in ['Season', 'Week', 'Home', 'Away', 'Network']:
    df[col].value_counts().plot(kind='bar')
    plt.show()
```
